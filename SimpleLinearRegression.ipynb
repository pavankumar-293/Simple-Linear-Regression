```python
# Step 1: Importing library and function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 2: Creating a dataset
data = {
    'Experience': [1, 2, 3, 4, 5, 4, 2, 3],
    'Education_Level': [1, 2, 1, 2, 3, 1, 3, 2],  
    'Age': [22, 25, 28, 30, 35, 30, 27, 26],
    'Salary': [30000, 35000, 50000, 60000, 65000, 65000, 32000, 55000]
}
df = pd.DataFrame(data)
df
```

```python
# Step 3: Independent variable (X) and dependent variable (y)
X = df[['Experience','Education_Level','Age']]
y = df[['Salary']]
```

```python
# Step 4: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2529)
```

```python
# Step 5: Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 6: Coefficients and Intercept
model.intercept_ , model.coef_
```

```python
# Step 7: Predict values for the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Step 8: Calculate Mean Squared Error and R-squared
mean_squared_error(y_test, y_pred), r2_score(y_test, y_pred)
```

```python
plt.scatter(X_test['Experience'], y_test, color='blue', label='Actual')
plt.scatter(X_test['Experience'], y_pred, color='red', label='Predicted')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Actual vs Predicted Salary')
plt.legend()
plt.show()
```
